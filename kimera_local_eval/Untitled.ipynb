```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Timestamp_Investigation - Sheet1.csv')
```

```python
# importing package
import matplotlib.pyplot as plt

filtered = df[df['Gyroscope Timestamp (ms)'] != 0]
filtered2 = df[df['Accelerometer Timestamp (ms)'] != 0]

# plot line
plt.scatter(filtered['Timeline (ms)'], filtered['Gyroscope Timestamp (ms)'])
plt.scatter(filtered2['Timeline (ms)'], filtered2['Accelerometer Timestamp (ms)'])
plt.show()
```
